```python
import json
from typing import List
import requests as r
import base64
from PIL import Image
from io import BytesIO
import json

ENDPOINT_URL = "https://sb6zn2j49353q6uf.us-east-1.aws.endpoints.huggingface.cloud"
HF_TOKEN = "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"

# helper image utils
def encode_image(image_path):
  with open(image_path, "rb") as i:
    b64 = base64.b64encode(i.read())
    result = b64.decode("utf-8")
  # print("b64.decode('utf-8'):", result)
  return result


def predict(prompt, image, mask_image):
    image = encode_image(image)
    mask_image = encode_image(mask_image)

    # prepare sample payload
    request = {"inputs": prompt, "image": image, "mask_image": mask_image}
    # headers
    headers = {
    "Authorization": f"Bearer {HF_TOKEN}",
    "Content-Type": "application/json",
    "Accept": "application/json"
    }

    response = r.post(ENDPOINT_URL, headers=headers, json=request)

    img_dict = json.loads(response.content.decode())
    images = []
    for img_str in img_dict.values():
      base64_image = base64.b64decode(img_str)
      buffer = BytesIO(base64_image)
      image = Image.open(buffer)
      images.append(image)

    return images
```

```python
prompt="A yellow Retriever dog, high resolution, sitting natrually on a park bench"
image="dog.png"
mask_image="dog_mask.png"

output_images = predict(prompt, image, mask_image)
# output_images[0]
```
